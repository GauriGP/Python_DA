```python
plt.figure(figsize=(10,6))
plt.title('Numerical Columns Analysis')
sns.kdeplot(x = df['Rating'], shade = True, color='r')
plt.xlabel('Rating')
plt.tight_layout()
plt.show()
```

### Top 10 categories 

```python
category_count = pd.DataFrame(df['Category'].value_counts())
```

```python
category_count.head(10)
```

```python
category_count.index[:10]
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=category_count.index[:10], y='count', data = category_count[:10], palette='hls')
plt.title('Top 10 Categories with installs')
plt.xticks(rotation=90)
plt.show()
```

### Top 5 catogories based on installs

```python
df.head()
```

```python
top_5_categories = pd.DataFrame(df.groupby('Category')['Installs'].sum().sort_values(ascending=False).head(5))
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Category',y='Installs',data=top_5_categories,palette='hls')
plt.title('Top 5 catogories based on installs')
plt.tight_layout()
plt.show()
```
