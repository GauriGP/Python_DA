```python
for i in df['Last Updated']:
    print(i.split(' ')[0])
```

```python
def month(x):
    return x.split(' ')[0]
```

```python
df['Month'] = df['Last Updated'].apply(month)
```

```python
df.head()
```

```python
def day(x):
    return x.split(' ')[1]
```

```python
df['Day'] = df['Last Updated'].apply(day)
```

```python
def year(x):
    return x.split(' ')[2]
```

```python
df['Year'] = df['Last Updated'].apply(year)
```

```python
df['Day'] = df['Day'].str.replace(',','')
```
